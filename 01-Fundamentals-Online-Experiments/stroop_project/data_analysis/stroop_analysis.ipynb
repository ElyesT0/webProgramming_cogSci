```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path_processed="/Users/elyestabbane/Documents/2-PhD/4-Teaching/programming_online_experiments/01-Fundamentals-Online-Experiments/data_analysis/data/processed/aggregated_data_stroop.json"
df=pd.read_json(path_processed)
```

```python
# Extract response times based on condition
rt_matching = df[df['condition'] == 'matching']['response_RT']
rt_non_matching = df[df['condition'] != 'matching']['response_RT']

# Compute mean response times
mean_rt_matching = rt_matching.mean()
mean_rt_non_matching = rt_non_matching.mean()

sem_rt_matching = np.std(rt_matching, ddof=1) / np.sqrt(len(rt_matching))
sem_rt_non_matching = np.std(rt_non_matching, ddof=1) / np.sqrt(len(rt_non_matching))

# Define categories
categories = ['Color-word\nMatching', 'Not Matching']
means = [mean_rt_matching, mean_rt_non_matching]
sems = [sem_rt_matching, sem_rt_non_matching]

# Plot the mean response times with error bars
plt.figure(figsize=(6, 4))
bar_positions = np.arange(len(categories))

# Bar plot
plt.bar(bar_positions, means, yerr=sems, capsize=5, color=['#219ebc', '#780000'], alpha=0.7, label="Mean ± SEM")

# Add random jitter to x-axis positions
jitter_matching = np.random.normal(loc=0, scale=0.05, size=len(rt_matching))
jitter_non_matching = np.random.normal(loc=0, scale=0.05, size=len(rt_non_matching))

# Scatter plot with jittered x-values
plt.scatter(bar_positions[0] + jitter_matching, rt_matching, color='#219ebc', alpha=0.5, label="Data points")
plt.scatter(bar_positions[1] + jitter_non_matching, rt_non_matching, color='#780000', alpha=0.5)

# Labels and title
plt.xticks(bar_positions, categories)
plt.ylabel('RT \n(ms)',rotation=0,fontsize=10, labelpad=20)
plt.title('Comparison of Response Time: Matching vs Not Matching')
plt.show()
```

```python
# Extract response times based on condition
total_trials_matching=len(df[df['condition']=='matching'])
total_trials_non_matching=len(df[df['condition']!='matching'])

success_matching = len(df[(df['condition']=='matching')&(df['performance']=='success')])
success_non_matching = len(df[(df['condition']!='matching')&(df['performance']=='success')])

# Compute mean response times
success_rate_matching = success_matching/total_trials_matching
success_rate_non_matching = success_non_matching/total_trials_non_matching

# Plot the mean response times with error bars
plt.figure(figsize=(6, 4))
bar_positions = np.arange(len(categories))

# Bar plot
plt.bar(bar_positions, [success_rate_matching,success_rate_non_matching], color=['#219ebc', '#780000'], alpha=0.7, label="Success rate")

# Labels and title
plt.xticks(bar_positions, categories)
plt.ylabel('Success\nRate',rotation=0,fontsize=10, labelpad=20)
plt.title('Comparison of Success Rates: Matching vs Not Matching')
plt.show()
```
